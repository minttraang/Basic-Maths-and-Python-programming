```python
# Implementing append() method using len() and index assignment
def my_append(lst, value):
    lst += [None] #increase the list size by 1
    '''
    The += operator with lists combines two lists together. 
    Adding [None] => creating a temporary "empty seat" or placeholder at the very end of your list.

    Before: [10, 20] (Length is 2)
    After this line: [10, 20, None] (Length is now 3)

    Now the list has expanded by exactly one slot, but that slot is currently occupied by None.
    '''
    lst[len(lst) - 1] = value
    '''
    This line replaces that placeholder None with your actual target value (which is 30).
    To target the very last slot, Python calculates the index using len(lst) - 1:
    len(lst) counts the new total number of items, which is 3.
    Because Python indexing starts at 0, the indexes of our list [10, 20, None] are:
        Index 0: 10
        Index 1: 20
        Index 2: None
    So, len(lst) - 1 becomes 3 - 1 = 2. Index 2 points precisely to our None slot!
    Finally, lst[2] = 30 overwrites the None with 30.
    Final List: [10, 20, 30]
    '''

#test
data = [6,5,7,1,9,2]
print(data)
my_append(data, 4)
print(data)
```

```python
#custom summation - way 1
def computeSum(data):
    result = 0

    for value in data:
        result = result + value
    return result

# test
data = [6, 5, 7, 1, 9, 2]
summation = computeSum(data)
print(summation)

#custom summation - way 2
def computeSummation(data):
    result = 0

    length =len(data)
    for index in range(length):
        result = result + data[index]
    return result
data = [6, 5, 7, 1, 9, 2]
summation = computeSummation(data)
print(summation)
```

```python
#sum of even number
def sum1(data):
    result = 0

    for value in data:
        if value %2 == 0:
            result += value
    return result

data = [6, 5, 7, 1, 9, 2]
summation = sum1(data)
print(summation)
```

```python
# sum of numbers with even indices
def sum2(data):
    result = 0

    length = len(data)
    for index in range(length):
        if index %2 == 0:
            result = result + data[index]
    return result

data = [6, 5, 7, 1, 9, 2]
summation = sum2(data)
print(summation)
```

# **LIST COMPREHENSION**

```python
#square function
def square_function(data):
    result = []

    for value in data:
        result.append(value*value)
    return result


# test
data = [6, 5, 7, 1, 9, 2]
print(data)
data_s = square_function(data)
print(data_s)

#using list comprehension
def square_function(data):
    result = [value*value for value in data]
    return result
data = [6, 5, 7, 1, 9, 2]
print(data)
data_s = square_function(data)
print(data_s)
```

```python
import math

#sigmoid function
def sigmoid(x):
    result = 1 / (1 + math.exp(-x))
    return result

def sigmoidForList(data):
    result = [sigmoid(x) for x in data]
    return result

# test
data = [1, 5, -4, 3, -2]
print(data)
data_a = sigmoidForList(data)
print(data_a)
```

```python
#relu function
def relu(x):
    result = 0
    if x > 0:
        result = x
    return result

def reluForList(data):
    result = [relu(x) for x in data]
    return result

# test
data = [1, 5, -4, 3, -2]
print(data)
data_a = reluForList(data)
print(data_a)

def relu2(data):
    result = [x if x > 0 else 0 for x in data]
    return result
# test
data = [1, 5, -4, 3, -2]
print(data)
data_a = relu2(data)
print(data_a)
```

# **MUTABLE & IMMUTABLE**

```python
#immutable
def square(data):
    result = []

    for value in data:
        result.append(value*value)
    return result
# test
data = [6, 5, 7, 1, 9, 2]
print(data)

data_s = square(data)
print(data_s)
```

```python
#mutable
def square(data):
    length = len(data)

    for i in range(length):
        value = data[i]
        data[i] = value*value

# test
data = [6, 5, 7, 1, 9, 2]
print(data)

square(data)
print(data)
```

# **INTEGRAL**

```python
import random

def GenList(a,b):
    result = []
    for _ in range(10):
        num = random.randint(a,b)
        result.append(num)
    return result

data = GenList(8,9)
print(data)
```

```python
def accumulate(data):
    result = []
    result.append(data[0])

    length = len(data)
    for i in range(1, length):
        result.append(data[i] + result[i-1])
    return result
result = accumulate(data)
print(result)
```
