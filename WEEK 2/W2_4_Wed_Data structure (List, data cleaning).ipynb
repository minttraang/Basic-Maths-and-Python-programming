# **2. LIST EXAMPLE**

```python
#create a list that have int variables

l = [1,2,3,4,5,6,7,8,9]

#call the variable using forward index
print(l[0])
print(l[1])

#call the variable using backward index
print(l[-1])
print(l[-2])
```

```python
n1 = 5
n2 = 8

l1=[n1]
l2 = [n1,n2]

print(l1)
print(l2)
```

```python
#ways to create a list

#range([start,] stop [,step])
#start and step is optional, meaning we do not have to assign value to the range() function
#default value of start is 0 and stap is 1

#create a blank list
l = []

#with each value of i, create a list contains 1 variable and append to list l
for i in range(10):
    l += [i]

#with each value of i, insert i into list l
for i in range(100):
    l.append(i)

#this is widely used
l = [i for i in range(1000)]

#force 'range' to be list
l = list(range(1000))
```
