```python
import math

def quad_equation(a, b, c):
    '''
    This function aims at solving the quadratic equation
    a, b, c --- three parameters and a =! 0
    '''

    #compute delta
    delta = b*b - 4*a*c

    if delta < 0:
        return()
    elif delta == 0:
        x = (-b + math.sqrt(delta))/2*a
        return (x,)
    else:
        x1 = (-b + math.sqrt(delta))/2*a
        x2 = (-b - math.sqrt(delta))/2*a
        return(x1,x2)
    
result = quad_equation(5,0,1)
print(type(result))
print(len(result))
print(result)

result = quad_equation(5,5,1)
print(type(result))
print(len(result))
print(result)

result = quad_equation(4,4,1)
print(type(result))
print(len(result))
print(result)
```

1. The Context: What is list3?
Before looking at lambda, notice that list3 pairs the letters and numbers together using zip(). It looks like this:

list3 = [('a', 16), ('g', 13), ('e', 18), ('h', 11), ('b', 15)]

2. Breaking Down the Lambda
When you want to sort list3, Python needs to know: “Should I sort by the letters, or should I sort by the numbers?”

That's where key=lambda item: item[1] comes in:

item: Represents each tuple in your list (e.g., ('a', 16)).
