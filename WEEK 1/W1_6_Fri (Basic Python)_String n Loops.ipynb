```python
import random

#Total times of coin flips
total_flips = 0

#Total times tail appears
total_tail = 0

#Total times head appears
total_head = 0

for _ in range(1000):
    #generate a random number in range[0,1)
    n = random.random()
    if n < 0.5:
        total_tail += 1
    else:
        total_head += 1
    
    #this line code does not belong to else block
    #each time flip, either head or tail will count to total flips
    total_flips += 1

print("total_flips: %d" % total_flips)
print('total_tail: %d' % total_tail)
print('total_head: %d' % total_head)
```

```python
#calculate factorial
def factorial(fact):
    result = 1

    for i in range(2, fact + 1):
        result *= i

    return result
print(factorial(9))
print(factorial(3))
print(factorial(5))
print(factorial(7))
print(factorial(8))
```

```python
#estimate e number
def estimate_e(n):
    result = 1

    for i in range(1, n + 1):
        result += 1/factorial(i)

    return result
#estimate e with n = 10
print(estimate_e(10))
print(estimate_e(40))
```

```python
#calculate square root using Babylonian method
def compute_sqrt(N, num_loops):
    '''
    This function aims to compute square root for the number N
    
    N -- the number needs to take the square root
    num_loops -- number of loops used for this optimization
    '''
    x_n = N/2.0
    print(x_n)

    for i in range(num_loops):
        x_np = (x_n + N/x_n) / 2.0
        x_n = x_np
        print(x_np)
    return x_np
print(compute_sqrt(N=9,num_loops=5))
print(compute_sqrt(N=16,num_loops=5))
```

**PI NUMBER ESTIMATION**

```python
# Gregory-Leibniz Series

n = 10000
PI = 0
for i in range(1, n + 1):
    PI = PI + (-1)**(i+1) / (2*i - 1)
PI *= 4

print("Estimated PI is: ", PI)
```

```python
# Nilakantha Series
n = 10
PI = 0
for i in range(n):
    PI = PI + (-1)**i / ((2*i+2)*(2*i+3)*(2*i+4))
PI = 3 + 4*PI
print("Estimated PI is: ", PI)
```

```python
import math

#No. points in a circle
N_T = 0
N = 100000

#Generate a random point N
for i in range(N):
    #generate x, y in range [-1,1]
    x = random.random() * 2 - 1
    y = random.random() * 2 - 1

    x2 = x**2
    y2 = y**2

    #check if p is inside the circle
    if math.sqrt(x2 + y2) <= 1.0:
        N_T += 1

#calculate PI
pi = (N_T / N) * 4
print(pi)
```

**LIMIT_FOR**

```python
def compute_y(x):
    return math.sqrt(1 - x*x)

delta_x = 0.01
n = int(2/ delta_x)

x = -1.0
half_area = 0.0
for _ in range(n):
    y = compute_y(x)
    half_area += delta_x*y
    x += delta_x

print(half_area*2)
```
