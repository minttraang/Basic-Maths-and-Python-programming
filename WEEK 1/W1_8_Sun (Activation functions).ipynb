## **Exercise 1 - Functions to calculate F1_score**
Viết hàm thực hiện đánh giá F1-Score cho các mô hình phân loại.
- $\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}$
- $\text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}$
- $\text{F1-score} = 2 * \frac{\text{Precision} * \text{Recall}}{\text{Precision} + \text{Recall}}$

- Input: nhận 3 giá trị **tp, fp, fn**

- Output: trả về kết quả của **F1-score**


**NOTE: Đề bài yêu cầu các điều kiện sau**
    
- Phải **kiểm tra giá trị nhận vào tp, fp, fn là kiểu dữ liệu int**, nếu là kiểu dữ liệu khác thì in thông báo cho người dùng ví dụ kiểm tra fn là float, print **'fn must be int'** và thoát hàm hoặc dừng chương trình.
- Yêu cầu **tp, fp, fn phải đều lớn hơn hoặc bằng 0**, nếu khác thì in **'tp and fp and fn must be greater than or equal zero'** và thoát hàm hoặc dừng chương trình. **Cần kiểm tra trường hợp phép tính chia 0.**

```python
import math
def calculate_f1_score(tp, fp, fn):
    '''
    Calculate the F1-score for a classification task.

    Args:
        tp (int): true positives, >= 0
        fp (int): fales positive, >= 0
        fn (int): false negative, >= 0

    Formula:
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f1_score = 2 * (precision * recall) / (precision + recall)

    Return:
        f1_score(float): the F1_score value,
        or None if the input is invalid
    '''
#1. Validate types (reject bool because bool is a subclass of int)
    for name, val in [('tp', tp), ('fp', fp), ('fn', fn)]:
        if isinstance(val, bool) or not isinstance(val, int):
            print(f'{name} mus be int')
            return None

#2. Validate non negative values
    if tp < 0 or fp < 0 or fn < 0:
        print('tp and fp and fn must be greater than or equal to zero')
        return None

#3. Calculate precision, Recall, F1_score (handle division by zero)
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0
    return f1_score

#Test
assert round(calculate_f1_score(tp = 2, fp = 3, fn = 5), 2) == 0.33
print(round(calculate_f1_score(tp=2, fp=4, fn=5), 2))

```

# **Exercise 2: Activation functions**

- Input:
    - Người dùng nhập giá trị **x**
    - Người dùng nhập tên **activation function chỉ có 3 loại (sigmoid, relu, elu)**
- Output: Kết quả **f(x)** (x khi đi qua activation function tương ứng theo activation function name). Ví dụ **nhập x=3, activation_function = 'relu'. Output: print 'relu: f(3)=3'**

**NOTE: Lưu ý các điều kiện sau:**
- Dùng function **is_number** được cung cấp sẵn để **kiểm tra x có hợp lệ hay không** (vd: x='10', is_number(x) sẽ trả về True ngược lại là False). Nếu **không hợp lệ print 'x must be a number' và dừng chương trình.**
- Kiểm tra **activation function name có hợp lệ hay không nằm trong 3 tên (sigmoid, relu, elu)**. **Nếu không print 'ten_function_user is not supported'** (vd người dùng nhập 'belu' thì print 'belu is not supported')
- Convert **x** sang **float** type
- Thực hiện với theo công thức với activation name tương ứng. Print ra kết quả
- Dùng math.e để lấy số e
- $\alpha$ = 0.01

```python
import math

def is_number(n):
    '''
    Check whether n is a valid number by trying to cast it to float. 
    Return True if successful, False otherwise 
    '''

    try:
        float(n)
    except ValueError:
        return False
    return True

assert is_number(3) is True
assert is_number('-2a') is False
print(is_number(1))
print(is_number('n'))
```

```python
x = -2.0
if x <= 0:
    y = 0.0
else:
    y = x
print(y)
```

```python
def calculate_relu(x):
    '''
    Calculate the ReLU functinos:
        ReLU(x) = max(0, x)
    
    Args:
        x (float): the input value

    Returns:
        result (float): the value of ReLU(x)
    '''

    if x <= 0:
        result = 0.0
    else:
        result = x
    return float(result)

assert calculate_relu(-3) == 0.0
assert calculate_relu(5) == 5.0
print(calculate_relu(5))
print(calculate_relu(10))
```

```python
import math
def calculate_sigmoid(x):
    '''
    Calculate the Sigmoid function:
        sigmoid(x) = 1 / (1 + e^(-x))

    Uses math.exp() instead of math.e ** (-x) for better numerical stability.

    Args:
        x (float): the input value

    Returns:
        float: the value of sigmoid(x) 
    '''

    return 1.0 / (1.0 + math.exp(-x))

assert round(calculate_sigmoid(3),2) == 0.95
print(round(calculate_sigmoid(2),3))
print(round(calculate_sigmoid(2),2))
```

```python
import math

def calculate_elu(x):
    '''
    Calculate the ELU (Exponential Linear Unit) function:
        ELU(x) = x if x >= 0
        ELU(x) = alpha * (e^x - 1) if x < 0

    NOte: this exercise uses alpha = 0.01 as required by the problem.

    Args:
        x (float): the input value
    
    Returns:
        result (float): the value of ELU(x)
    '''

    alpha = 0.01
    if x < 0:
        result = alpha * (math.exp(x) - 1)
    else:
        result = x
    return result

assert round(calculate_elu(1)) == 1
print(round(calculate_elu(-1),2))
```

```python
import math

def calculate_activation_function(x, act_name):
    '''
    Calculate the activation function for x based on act_name.

    Args:
        x (float): the input value.
        act_name (str): name of the activation function, one of 'relu', 'sigmoid', or 'elu'.capitalize

    Returns:
        float: the value of f(x) corresponding to act_name, or None if act_name is invalid.
    '''

    if act_name == 'relu':
        return calculate_relu(x)
    if act_name == 'sigmoid':
        return calculate_sigmoid(x)
    if act_name == 'elu':
        return calculate_elu(x)
    return None

assert calculate_activation_function(x=1, act_name='relu') == 1.0
print(round(calculate_activation_function(x=3, act_name='sigmoid'),2))
```

```python
def interactive_activation_function():
    x = input('Input x = ')
    if not is_number(x):
        print('x must be a number!!')
        return None
    
    act_name = input('Input activation function (sigmoid | relu | elu): ')
    x = float(x)
    result = calculate_activation_function(x, act_name)
    if result is None:
        print(f'{act_name} is nor supported')
    else:
        print(f'{act_name}: f({x}) = {result}')
    

interactive_activation_function()
```

# **Exercise 3: Trigonometric functions**

Viết 4 functions để ước lượng các hàm số sau sin(x), cos(x), sinh(x), cosh(x).
-  Input: x (số muốn tính toán) và n (số lần lặp muốn xấp xỉ)
- Output: Kết quả ước lượng hàm tương ứng với x. Ví dụ hàm cos(x=0) thì output = 1

**NOTE: Các bạn chú ý các điều kiện sau**
- x là radian
- n là số nguyên dương > 0
- các bạn nên viết một hàm tính giai thừa riêng

```python
def factorial(x):
    '''
    Calculate the factorial of x (x! where x is a non-negative integer)
    '''

# The Python isinstance() function checks if an object belongs to a specific data type or class, returning either True or False. 
# It is widely used for type validation to ensure that code handles data safely.
    if not isinstance(x, int):
        raise ValueError('x must be a non-negative integer') #When you write raise ValueError, you are telling Python to manually crash and stop your program because a strict rule was broken.
    
    res = 1
    for i in range(1, x + 1):
        res *= i
    return res

assert factorial(0) == 1
assert factorial(4) == 24
print(factorial(8))
```

```python
import math

def approximate_sin(x, n):
    '''
    Approximate sin(x) using a Taylor series with n+1 terms

    Args:
        x (float): angle in radians
        n (int): positive integer, number of iterations

    Returns:
        sin_approx (float): the approx value of sin(x)
    '''

    if not isinstance(n, int) or isinstance(n, bool) or n <= 0:
        raise ValueError('n must be a positive integer')
    
    sin_approx = 0
    for i in range(n + 1):
        coef = (-1) ** i
        num = x ** (2 * i + 1)
        denom = factorial(2 * i + 1)
        sin_approx += coef * (num/denom)
    return sin_approx

assert round(approximate_sin(x=1, n=10), 4) == round(math.sin(1),4)
print(round(approximate_sin(x=3.14, n=10),4))
```

```python
def approximate_cos(x, n):
    '''
    Approximate cos(x) using a Taylor series with n+1 terms

    Args:
        x (float): angle in radians
        n (int): positive integer, number of iterations

    Returns:
        cos_approx (float): the approx value of cos(x)
    '''

    if not isinstance(n, int) or isinstance(n, bool) or n <= 0:
        raise ValueError('n must be a positive integer')
    
    cos_approx = 0
    for i in range(n+1):
        coef = (-1) ** i
        num = x ** (2 * i)
        denom = factorial(2*i)
        cos_approx += coef * (num/denom)
    return cos_approx

assert round(approximate_cos(x=1, n=10), 2) == 0.54
print(round(approximate_cos(x=3.14, n=10),2))
```

```python
def approximate_sinh(x, n):
    """
    Approximate sinh(x) using a Taylor series with n+1 terms.

    Args:
        x (float): the input value.
        n (int): positive integer, number of iterations.

    Returns:
        sinh_approx (float): the approximate value of sinh(x).
    """
    if not isinstance(n, int) or isinstance(n, bool) or n <= 0:
        raise ValueError('n must be a positive integer')
    
    sinh_approx = 0
    for i in range(n+1):
        num = x ** (2 * i + 1)
        denom = factorial(2 * i + 1)
        sinh_approx += num/denom
    return sinh_approx

assert round(approximate_sinh(x=1, n=10), 2) == 1.18
print(round(approximate_sinh(x=3.14, n=10), 2))
```

**Lưu ý:** Bản này sử dụng type hints để chú thích kiểu dữ liệu trong Python:

- Cú pháp: `def func(x: float, n: int) -> float:`
    - `x: float` — chú thích tham số `x` là kiểu `float`
    - `n: int` — chú thích tham số `n` là kiểu `int`
    - `-> float` — chú thích giá trị trả về là `float`
- Type hints **không bắt buộc** và **không ép kiểu lúc chạy** — chỉ là gợi ý cho IDE, công cụ kiểm tra (mypy, pyright), và người đọc code.

```python
def approximate_cosh(x: float, n: int) -> float:
    """
    Approximate cosh(x) using a Taylor series with n+1 terms.

    Args:
        x (float): the input value.
        n (int): positive integer, number of iterations.

    Returns:
        cosh_approx (float): the approximate value of cosh(x).
    """
    if not isinstance(n, int) or isinstance(n, bool) or n <= 0:
        raise ValueError('n must be a positive integer')
    
    cosh_approx: float = 0.0
    for i in range(n + 1):
        num: float = x ** (2 * i)
        denom: int = factorial(2*i)
        cosh_approx += num / denom
    return cosh_approx

assert round(approximate_cosh(x=1, n=10), 2) == 1.54
print(round(approximate_cosh(x=3.14, n=10), 2))
```
