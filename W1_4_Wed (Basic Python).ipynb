```python
#SIGMOID?

import math

def compute_sigmoid(x):
    '''
    this function aims to compute the sigmoid value of a number
    
    x -- input number
    
    Output is sigmoid value
    '''

    result_sig = 1 / (1 + math.exp(-x))
    return result_sig

#testing
x = 3.0
result = compute_sigmoid(x)
print(result)
```

**4. UNIT CIRCLE**

Compute the area of a unit circle

```python
def compute_area_circle(selected_degree):
    #sin and cosine value of the selected_degree
    sin_degree = math.sin(math.radians(selected_degree))
    cos_degree = math.cos(math.radians(selected_degree))

    #compute the area for the part
    area_degree = sin_degree * cos_degree / 2

    #summarise (360/selected_degree) parts
    area = area_degree * (360/selected_degree)

    return area

#testing
area = compute_area_circle(0.02)
print(area)
```

**6. MATCH-CASE_TYPE-HINT**

```python
#MATCH CASE

num = 3

match num:
    case 1:
        print("One")
    case 2:
        print('Two')
    case 3:
        print('Three')
    case _: 
        print("Else")
```

```python
#TYPE HINT_(also known as a type annotation) in Python.
'''
num: This is the name of the variable.

: int: This is the type hint. It explicitly states that the variable num is expected to hold an integer (a whole number).

= 5: This assigns the integer value 5 to the variable.
'''

num: int = 5
print(num)
```

**7. RULE-BASED CHATBOT (simple implementation)**

```python
def respond_to_user(user_input):
    #process the input from the users
    user_input = user_input.lower().strip() #makes all lowercase and strip removes any leading, and trailing whitespaces

    #Say hello to user
    if user_input == 'hi':
        print(
            "Hello customer! How can I assist you today?\n" #\n == new line
            "['Advise on products', 'Warranty search', 'Technical assistant]"
        )
    elif user_input == "hello":
        print(
            "Hello customer! Can I help you with anything today?\n" #\n == new line
            "['Advise on products', 'Warranty search', 'Technical assistant']"
        )

    #Advise on products
    elif user_input == "advise on products":
        print(["macbook", "ipad", "iphone"])

    elif user_input == "macbook":
        print("Which type of macbook do you want to buy today?")

    elif user_input == "ipad":
        print("Which type of ipad do you want to buy today?")

    elif user_input == "macbook":
        print("Which type of macbook do you want to buy today?")

    elif user_input == "iphone":
        print("Which type of iphone do you want to buy today?")

    #Warranty search
    elif user_input == "warranty search":
        print(["Search via mobile number", "Search via email"])

    elif user_input == "Search via mobile number":
        print("Please input your registered mobile number: ")

    elif user_input == "Search via email":
        print("Please input your registered email: ")

    #Technical assistant
    elif user_input == "technical assistant":
        print(["Hardware error", 'Software error'])
    elif user_input == "Hardware error":
        print("Please describe the error to receive help from our assistant: ")
    elif user_input == "Software error":
        print("Please describe the error to receive help from our assistant: ")

    #DUMB and do not understand the customer's requirement
    else:
        print("Sorry! I do not understand your requirement. Can you please provide more details?")
```

```python
user_input = "hello"
respond_to_user(user_input)
```
